```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2
from norfair import Tracker,Detection
```

```python
model=YOLO('yolov8n.pt')
cap=cv2.VideoCapture('/content/ahmed.mp4')
```

```python
fram_hight=int(cap.get(3))
fram_width=int(cap.get(4))
fbs=int(cap.get(5))
out=cv2.VideoWriter('output.mp4',cv2.VideoWriter_fourcc(*'mp4v'),fbs,(fram_width,fram_hight))
```

```python
tracker=Tracker(distance_function='euclidean',distance_threshold=30)
person_counter=[]
heatmap=np.zeros((fram_hight,fram_width),dtype=np.float32)
allpoints=[]
fram_idx=0
while True:
    ret,frame=cap.read()
    if not ret:
       break

    results=model(frame)[0]
    detections=[]
    for result in results.boxes.data.tolist():
      x1,y1,x2,y2,score,class_id=result
      if (class_id==0)and score>.4:
        cx=int(x1+x2)/2
        cy=int(y1+y2)/2
        detections.append(Detection(points=np.array([[cx,cy]])))
    tracker_obj=tracker.update(detections=detections)
    person_counter.append(len(tracker_obj))
```

```python
for obj in tracker_obj:
  x,y=obj.estimate[0]
  allpoints.append((int(x),int(y)))
  x_clipped = np.clip(int(x), 0, fram_width - 1)
  y_clipped = np.clip(int(y), 0, fram_hight - 1)
  heatmap[y_clipped, x_clipped] += 1
  cv2.circle(frame, (int(x), int(y)), 5, (0, 255, 0), -1)
  cv2.putText(frame, f'ID {obj.id}', (int(x), int(y)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,0), 1)

    # حفظ الفريم
  out.write(frame)
  fram_idx += 1

cap.release()
out.release()
print("✅ تم حفظ الفيديو الناتج: output_tracking.mp4")

```

```python
plt.figure(figsize=(10, 6))
plt.imshow(heatmap, cmap='hot', interpolation='nearest')
plt.axis('off')
plt.title("Crowd Density Heatmap")
plt.savefig("heatmap_output.png")
plt.show()
print("✅ تم حفظ خريطة التزاحم heatmap_output.png")
```

```python
df = pd.DataFrame({
    "frame": list(range(len(person_counter))),
    "person_count": person_counter
})
df.to_csv("person_count.csv", index=False)
print("✅ تم حفظ الإحصائيات في person_count.csv")
```
